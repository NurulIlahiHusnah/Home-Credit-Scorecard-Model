```python
import time
cols = ['DAYS_BIRTH', 'DAYS_REGISTRATION', 'DAYS_EMPLOYED']
tahun_2018 = pd.to_datetime('2018-12-31')

for col in cols:
    train[col] = tahun_2018 + pd.to_timedelta(train[col], 'D')

# membuat feature baru penganti feature pada cols
train['AGE']= ((tahun_2018 - train['DAYS_BIRTH']).dt.days/365).round(0)
train['REGISTRATION_TIME']= ((tahun_2018 - train['DAYS_REGISTRATION']).dt.days/365).round(0)
train['WORK_DURATION']= ((tahun_2018 - train['DAYS_EMPLOYED']).dt.days/365).round(0)

# menghapus feature yang tidak diperlukan
train.drop(columns= cols, inplace=True)
print("Data setelah diubah:")
train.sample()
```

```python
def remove_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    df_cleaned = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_cleaned

df_cleaned_heatmap = remove_outliers_iqr(data)
df_cleaned_rfe = remove_outliers_iqr(data_rfe)
```

```python
# @title Hyperparameter Tuning (Nested Cross Validation)
# Untuk menentukan nilai cv optimal
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV

X_train = df_scale_train.drop(columns='TARGET') # saya menggunakan data yang sudah dilakukan standarisasi
y_train = df_scale_train['TARGET']

X_test = df_scale_test

# Inisialisasi Model dan pipeline
pipeline = Pipeline([
    ('model', LogisticRegression())
])

# Menentukan Parameter grid untuk mencari hyperparameter optimal
param_grid ={
    'model__C':[0.01, 0.1, 1, 10, 100],
    'model__penalty':['l1','l2'],
    'model__solver':[ 'liblinear']
}

# Nilai cv yang akan saya uji
cv_values=[3, 5, 7, 10]
result = {}

for cv in cv_values:
  # Untuk inner loop cross validation
  inner_cv = KFold(n_splits=cv, shuffle=True, random_state=42)
  random_search = RandomizedSearchCV(pipeline, param_distributions=param_grid, n_iter=50, cv=inner_cv, scoring='accuracy', n_jobs=-1, random_state=42)
  # Untuk outer loop cross validation
  outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)
  nested_scores = cross_val_score(random_search, X_train, y_train, cv=outer_cv)

# Hasil rata-rata skor
result[cv]= np.mean(nested_scores)
print(f"Mean accuracy for cv={cv}: {np.mean(nested_scores)}")

# Menampilkan nilai cv terbaik
optimal_cv = max(result, key=result.get)
print(f"Optimal cv: {optimal_cv} with mean accuracy: {results[optimal_cv]}")

```
